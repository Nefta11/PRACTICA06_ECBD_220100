## 5. Estadisticas Avanzada 

#### Calcular la medida de tendencias  central, de nuestro dataFrame

```python
import pandas as pd
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_statistics(df):
    stats = {}
    
    # Seleccionamos solo columnas numéricas
    numeric_cols = df.select_dtypes(include=[np.number]).columns

    if numeric_cols.empty:
        print("No hay columnas numéricas en el DataFrame.")
        return pd.DataFrame()

    for col in numeric_cols:
        series = df[col].dropna()  # eliminamos valores nulos

        # Evitar errores si la columna está vacía
        if series.empty:
            continue

        stats[col] = {
            'Mean': series.mean(),
            'Median': series.median(),
            'Mode': series.mode().iloc[0] if not series.mode().empty else np.nan,
            'Range': series.max() - series.min(),
            'Stdev': series.std(),
            'Variance': series.var(),
            'IQR': series.quantile(0.75) - series.quantile(0.25),
            '25%': series.quantile(0.25),
            '50%': series.quantile(0.50),
            '75%': series.quantile(0.75),
            'Min': series.min(),
            'Max': series.max(),
            'Skewness': skew(series),
            'Kurtosis': kurtosis(series)
        }

    # Crear DataFrame y transponerlo
    stats_df = pd.DataFrame(stats).T.round(2)
    return stats_df

# USO
stats_df = calculate_statistics(df_vacios_elimnados)
print(stats_df)  # Asegúrate de ver el resultado
```

```python
# Verificar valores nulos
df_api.isnull().sum()
```

```python
# Eliminar valores nulos
df_api_clean = df_api_clean.dropna()

df_api_clean.shape
```

### Transformación de Datos

```python
# Aplicar transformaciones a los datos del API
df_api_clean['phone_format'] = df_api_clean['phone'].apply(
    lambda x: 'US_Format' if x.startswith('1-') or x.startswith('(') 
    else 'International' if x.startswith('0') 
    else 'Other'
)

df_api_clean['user_category'] = pd.cut(df_api_clean['id'], 
                                      bins=[0, 3, 7, float('inf')], 
                                      labels=['Nuevo', 'Intermedio', 'Avanzado'])

df_api_clean['first_name'] = df_api_clean['name'].str.split().str[0]
df_api_clean['last_name'] = df_api_clean['name'].str.split().str[-1]

df_api_clean.head()
```
